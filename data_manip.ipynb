```python
import pandas as pd


df = pd.read_excel("/Users/yusufmoola/Desktop/june27.xlsx", sheet_name= "Map")

map  = {}

for row in df.itertuples(index=False):
    map[row.video_file] = row.tipper_file_name
```

```python
import json 

# Specify the filename
filename = 'mapJune27.json'

# Dump the dictionary into the JSON file
with open(filename, 'w') as json_file:
    json.dump(map, json_file, indent=4) # Using 'indent' for human-readable formatting

print(f"Dictionary successfully dumped to {filename}")
```

```python
import os
import shutil

# Configuration
source_dir = "/Users/yusufmoola/Library/CloudStorage/OneDrive-UHN/Videos/2025-06-27/sub368"
dest_dir = "videos_renamed/2025-06-27/sub368"

# Example map dictionary
# Key: current file name, Value: new file name
video_map = map

# Create destination directory if it doesn't exist
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

# Iterate through files in the source directory
for filename in os.listdir(source_dir):
    # Process only .mp4 files that exist as keys in the dictionary
    if filename.endswith(".MP4") and filename in video_map:
        # Get the value from the dictionary (e.g., "renamed_file_A.mat")
        dict_value = video_map[filename]
        
        # Strip the .mat extension and replace it with .mp4
        # os.path.splitext returns a tuple (root, ext)
        base_name = os.path.splitext(dict_value)[0]
        new_filename = f"{base_name}.MP4"
        
        # Build full paths
        src_path = os.path.join(source_dir, filename)
        dst_path = os.path.join(dest_dir, new_filename)
        
        # Copy and rename simultaneously
        shutil.copy(src_path, dst_path)
        print(f"Copied: {filename} -> {new_filename}")

print("Task complete.")
```
