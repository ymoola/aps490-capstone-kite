## Video rename based on mapping

```python
import os
import shutil

# Configuration
source_dir = "/Users/yusufmoola/Library/CloudStorage/OneDrive-UHN/Videos/2025-06-27/sub368"
dest_dir = "videos_renamed/2025-06-27/sub368"

# Example map dictionary
# Key: current file name, Value: new file name
video_map = map

# Create destination directory if it doesn't exist
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

# Iterate through files in the source directory
for filename in os.listdir(source_dir):
    # Process only .mp4 files that exist as keys in the dictionary
    if filename.endswith(".MP4") and filename in video_map:
        # Get the value from the dictionary (e.g., "renamed_file_A.mat")
        dict_value = video_map[filename]
        
        # Strip the .mat extension and replace it with .mp4
        # os.path.splitext returns a tuple (root, ext)
        base_name = os.path.splitext(dict_value)[0]
        new_filename = f"{base_name}.MP4"
        
        # Build full paths
        src_path = os.path.join(source_dir, filename)
        dst_path = os.path.join(dest_dir, new_filename)
        
        # Copy and rename simultaneously
        shutil.copy(src_path, dst_path)
        print(f"Copied: {filename} -> {new_filename}")

print("Task complete.")
```

## Check if all videos renamed are present in mapping

```python
from pathlib import Path
import json

root = Path("/Users/yusufmoola/Library/CloudStorage/OneDrive-UHN/videos_renamed")
maps_dir = Path("rename_mappings")

missing_by_date = {}

for date_dir in sorted(p for p in root.iterdir() if p.is_dir()):
    map_path = maps_dir / f"map{date_dir.name}.json"
    if not map_path.exists():
        print(f"Missing map file: {map_path}")
        continue

    with map_path.open("r", encoding="utf-8") as f:
        data = json.load(f)

    # Normalize map values to filenames only (case-insensitive)
    map_values = {
        Path(v).name.lower()
        for v in data.values()
        if isinstance(v, str)
    }

    # Collect all video files under date/participant/*
    video_files = {}
    for participant_dir in (p for p in date_dir.iterdir() if p.is_dir()):
        for vid in participant_dir.iterdir():
            if vid.is_file() and vid.suffix.lower() == ".mp4":
                video_files[vid.name.lower()] = vid.name

    missing_keys = sorted(set(video_files) - map_values)
    if missing_keys:
        missing_by_date[date_dir.name] = [video_files[k] for k in missing_keys]

if not missing_by_date:
    print("All videos are present in their respective map JSON values.")
else:
    for date, files in missing_by_date.items():
        print(f"{date}: {len(files)} missing")
        for name in files:
            print(f"  - {name}")
```

# Rename Erroenous files post manual review

```python
import os
from pathlib import Path

# =========================
# CONFIGURATION
# =========================

BASE_DIR = Path("/Users/yusufmoola/Library/CloudStorage/OneDrive-UHN/Li, Yue (Sophia)'s files - videos_renamed")

# Rename table (Date, Filename without extension, Action)
RENAME_MAP = [
    ("2025-02-12", "idapt797_sub344_DF_12_13-45-20", "P"),
    ("2025-02-12", "idapt797_sub344_UF_12_13-43-34", "Remove"),
    ("2025-02-12", "idapt797_sub344_UF_13_13-42-54", "Remove"),
    ("2025-02-12", "idapt799_sub344_DP_9_13-14-33", "F"),
    ("2025-02-12", "idapt797_sub349_DF_13_12-33-45", "P"),
    ("2025-02-12", "idapt797_sub349_UF_15_12-30-55", "P"),
    ("2025-02-12", "idapt797_sub349_UP_14_12-31-51", "F"),
    ("2025-02-12", "idapt799_sub349_UP_12_11-54-18", "P"),
    ("2025-02-12", "idapt799_sub349_UP_13_11-53-27", "F"),
    ("2025-02-12", "idapt798_sub360_UP_11_15-01-48", "F"),
    ("2025-02-12", "idapt800_sub360_UP_7_15-21-51", "F"),
    ("2025-02-13", "idapt798_sub354_DP_11_14-20-21", "F"),
    ("2025-02-13", "idapt798_sub354_DU_0_14-15-14", "P"),
    ("2025-02-13", "idapt798_sub354_UP_11_14-19-59", "F"),
    ("2025-02-13", "idapt798_sub354_UU_0_14-14-37", "P"),
    ("2025-02-13", "idapt800_sub354_DF_12_14-44-51", "Remove"),
    ("2025-02-13", "idapt800_sub354_UU_0_14-33-49", "P"),
    ("2025-02-13", "idapt800_sub357_UP_13_12-52-48", "P"),
]

# Set to True if you want to actually perform changes
EXECUTE = True


# =========================
# CORE LOGIC
# =========================

def find_file_recursive(root_folder: Path, filename_base: str):
    """
    Searches recursively for filename_base.mp4 or .MP4
    """
    for path in root_folder.rglob("*"):
        if path.is_file():
            name_no_ext = path.stem
            if name_no_ext == filename_base:
                return path
    return None


def update_direction_result(filename_base: str, new_result: str):
    """
    Example:
    idapt797_sub344_DF_12_13-45-20
    -> change DF to DP if new_result = 'P'
    """
    parts = filename_base.split("_")

    # parts[2] should be direction+result (e.g., DF, UP, etc.)
    dir_res = parts[2]

    direction = dir_res[0]   # U or D
    # old_result = dir_res[1]

    new_dir_res = direction + new_result
    parts[2] = new_dir_res

    return "_".join(parts)


def process_files():
    for date, filename_base, action in RENAME_MAP:
        date_folder = BASE_DIR / date

        if not date_folder.exists():
            print(f"[WARNING] Date folder not found: {date_folder}")
            continue

        file_path = find_file_recursive(date_folder, filename_base)

        if not file_path:
            print(f"[NOT FOUND] {filename_base} in {date}")
            continue

        if action == "Remove":
            print(f"[DELETE] {file_path}")
            if EXECUTE:
                file_path.unlink()
            continue

        # Rename case
        new_filename_base = update_direction_result(filename_base, action)
        new_path = file_path.with_name(new_filename_base + file_path.suffix)

        print(f"[RENAME] {file_path.name} -> {new_path.name}")

        if EXECUTE:
            file_path.rename(new_path)

process_files()
```

# Trimming Longgg Vids

```python
import os
from pathlib import Path  
import cv2       

ROOT = Path("/Users/yusufmoola/Desktop/videos_renamed_360p")
MIN_DURATION_SEC = 20

def get_video_duration_sec(path):
    cap = cv2.VideoCapture(str(path))
    if not cap.isOpened():
        return None
    frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    if fps and fps > 0:
        return frame_count / fps
    return None

long_videos = []
for root, dirs, files in os.walk(ROOT):
    for f in files:
        if not f.lower().endswith(".mp4"):
            continue
        full_path = Path(root) / f
        duration = get_video_duration_sec(full_path)
        if duration is not None and duration > MIN_DURATION_SEC:
            long_videos.append(str(full_path.resolve()))

for p in sorted(long_videos):
    print(p)
print(f"\nTotal: {len(long_videos)} videos longer than {MIN_DURATION_SEC} seconds")
```

# Renaming subs from new folder

```python
from pathlib import Path

# =========================
# CONFIG
# =========================

TARGET_DIR = Path(
    "/Users/yusufmoola/Library/CloudStorage/OneDrive-UHN/Li, Yue (Sophia)'s files - new video and tipper files/Tipper files/2025-07-03"
)

SUB_RENAME_MAP = {
    "sub297": "sub295",
    "sub378": "sub374",
}

EXECUTE = True  


# =========================
# RENAME LOGIC
# =========================

for file_path in TARGET_DIR.iterdir():

    if not file_path.is_file():
        continue

    original_name = file_path.name
    new_name = original_name

    for wrong_sub, correct_sub in SUB_RENAME_MAP.items():
        if wrong_sub in new_name:
            new_name = new_name.replace(wrong_sub, correct_sub)

    # If name changed, attempt rename
    if new_name != original_name:
        new_path = file_path.with_name(new_name)

        if new_path.exists():
            print(f"[SKIP - TARGET EXISTS] {new_path}")
            continue

        print(f"[RENAME] {original_name} -> {new_name}")

        if EXECUTE:
            file_path.rename(new_path)
```
